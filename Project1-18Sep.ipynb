Objective
The goal of this study is to develop a machine learning model that can predict the quality of wine based on various physicochemical properties such as acidity, alcohol content, pH, and sulfur dioxide levels.

```python
#Read & view dataset
#Read the dataset dataset8 wine_quality type 1_with_missing.csv into Python and store it into a dataframe.
#Display the first 5 rows.

import pandas as pd

rawData = "dataset8 wine_quality type 1_with_missing.csv"
#db_wineQualityType = pd.read_csv(rawData)
rawData = pd.read_csv(rawData, dtype={'remaining_lease': 'string'})
#db_wineQualityType = pd.read_csv(rawData, low_memory=False)
display(rawData.head())
```

```python
# Check type of quality
variables_list = rawData['quality'].unique().tolist()

print("Variables in 'quality':", variables_list)
```

```python
# Re-arrange quality column as first column
cols = ['quality'] + [col for col in rawData.columns if col !='quality']
rawData = rawData[cols]
rawData.tail(10)
```

```python
# Number of occurrence of each quality
num_quality = rawData['quality'].value_counts()

print(num_quality)
```

```python
# Display all rows that contain NaN in their column
rawData.isnull()
rows_with_nulls = rawData[rawData.isnull().any(axis=1)]
rows_with_nulls.tail(5)
```

```python
# # Number of NaN values in each column
# print("NaN values in each column\n")
# for col in rawData.columns:
#     print(rawData.iloc[:, col].name, ":", rawData.iloc[:, col].isnull().sum(axis=0))

# Loop through all column and find which contain NaN
print("NaN values in columns")
for col in rawData.columns:
    num_nulls = rawData[col].isnull().sum()
    if num_nulls > 0:
        print(f"{col}: {num_nulls}")
```

```python
# Calculate mean|median|standard deviation with groupby quality type
import pandas as pd

# Group by quality and calculate mean, median, std for all numeric columns
wine_statistics = rawData.groupby('quality').agg(['mean', 'median', 'std']).round(2)

# Transpose so features are rows
wine_statistics_T = wine_statistics.T

# # Reset index
# wine_statistics_T = wine_statistics_T.reset_index()
# wine_statistics_T.columns = ['feature', 'stat'] + list(wine_statistics_T.columns[2:])

# # Display the table
# print(wine_statistics_T)

# # Display the table with segregation between each feature
# for feature in wine_statistics_T.index.levels[0]:
#     stats = wine_statistics_T.loc[feature]
#     print(f"Feature: {feature}")
#     print(stats)
#     print() # blank line after each feature

# Display the table by loop
for feature in wine_statistics_T.index.levels[0]:
    stats = wine_statistics_T.loc[feature]
    # Reset index so row headers become column headers
    stats_reset = stats.reset_index()
    stats_reset.columns = ['Stat'] + list(stats_reset.columns[1:])

    print(f"Feature: {feature}")
    from tabulate import tabulate
    print(tabulate(stats_reset, headers='keys', tablefmt='psql', stralign='left', numalign='left'))
```

```python
import pandas as pd
import os

# Columns with NaN
cols_with_nan = ['fixed acidity', 'citric acid', 'pH']

# DataFrames with different imputations
db_wineMean = rawData.copy()
db_wineMedian = rawData.copy()
db_wineStd = rawData.copy()

# Apply imputations
for col in cols_with_nan:
    db_wineMean[col] = db_wineMean.groupby('quality')[col].transform(lambda x: x.fillna(x.mean()))
    db_wineMedian[col] = db_wineMedian.groupby('quality')[col].transform(lambda x: x.fillna(x.median()))
    db_wineStd[col] = db_wineStd.groupby('quality')[col].transform(lambda x: x.fillna(x.std()))

# Store DataFrames and names in a dictionary for looping
imputed_dfs = {
    "Mean": db_wineMean,
    "Median": db_wineMedian,
    "Std": db_wineStd
}

# Loop to check NaNs
print("Check if NaN still exists.\n")
for name, df in imputed_dfs.items():
    print(f"{name} imputation NaN:")
    print(df[cols_with_nan].isna().sum())
    print()

    # File path to be saved
    file_name = f"wine_imputed_{name}.csv"
    # Create new file or overwrite if existed
    if os.path.exists(file_name):
        print(f"File {file_name} already existed and was overwritten.\n")
    else:
        print(f"File {file_name} created.\n")
```
